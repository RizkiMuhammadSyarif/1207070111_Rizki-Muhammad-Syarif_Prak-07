Load & Plot Input Image

```python
# Mengimpor library yang diperlukan
from skimage.io import imread
import matplotlib.pyplot as plt

# Membaca citra 1 dari file "Shikamaru.jpeg"
citra1 = imread(fname="Shikamaru.jpeg")

# Membaca citra 2 dari file "Random.jpg"
citra2 = imread(fname="Random.jpg")

# Mengubah citra1 dari BGR ke grayscale
citra1 = cv2.cvtColor(citra1, cv2.COLOR_BGR2GRAY)

# Mengubah citra2 dari BGR ke grayscale
citra2 = cv2.cvtColor(citra2, cv2.COLOR_BGR2GRAY)

# Menampilkan dimensi citra 1
print('Shape citra 1 : ', citra1.shape)

# Menampilkan dimensi citra 2
print('Shape citra 2 : ', citra2.shape)

# Membuat subplot dengan 1 baris dan 2 kolom
fig, axes = plt.subplots(1, 2, figsize=(10, 10))
ax = axes.ravel()

# Menampilkan citra 1 pada subplot pertama
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Citra 1")

# Menampilkan citra 2 pada subplot kedua
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Citra 2")
```

Menyiapkan variable output

```python
# Melakukan penyalinan citra pertama dan citra kedua
copyCitra1 = citra1.copy()
copyCitra2 = citra2.copy()

# Mengambil dimensi citra pertama
m1, n1 = copyCitra1.shape[:2]

# Membuat array kosong dengan ukuran yang sesuai untuk citra pertama
output1 = np.empty((m1, n1))

# Mengambil dimensi citra kedua
m2, n2 = copyCitra2.shape[:2]

# Membuat array kosong dengan ukuran yang sesuai untuk citra kedua
output2 = np.empty((m2, n2))

# Menampilkan informasi dimensi citra pertama
print('Shape copy citra 1 : ', copyCitra1.shape)
print('Shape output citra 1 : ', output1.shape)
print('m1 : ', m1)
print('n1 : ', n1)
print()

# Menampilkan informasi dimensi citra kedua
print('Shape copy citra 2 : ', copyCitra2.shape)
print('Shape output citra 2 : ', output2.shape)
print('m2 : ', m2)
print('n2 : ', n2)
print()
```

Proses Filter Median Pada Citra Input 1

```python
# Melakukan loop untuk setiap baris dalam rentang 1 hingga m1-1
for baris in range(1, m1-1):
    # Melakukan loop untuk setiap kolom dalam rentang 1 hingga n1-1
    for kolom in range(1, n1-1):
        # Mengambil data piksel sekitar dari citra 1 dan menyimpannya dalam list dataA
        dataA = [copyCitra1[baris-1, kolom-1], copyCitra1[baris-1, kolom], copyCitra1[baris-1, kolom+1], \
                 copyCitra1[baris, kolom-1], copyCitra1[baris, kolom], copyCitra1[baris, kolom+1], \
                 copyCitra1[baris+1, kolom-1], copyCitra1[baris+1, kolom], copyCitra1[baris+1, kolom+1]]
        
        # Mengurutkan dataA secara ascending
        dataA.sort()
        
        # Mengambil nilai median (elemen ke-4 setelah diurutkan) dan menyimpannya dalam output1
        output1[baris, kolom] = dataA[4]
```

Proses Filter Median Pada Citra Input 2

```python
# Melakukan loop untuk setiap baris dalam rentang 1 hingga m2-1
for baris in range(1, m2-1):
    # Melakukan loop untuk setiap kolom dalam rentang 1 hingga n2-1
    for kolom in range(1, n2-1):
        # Mengambil data piksel sekitar dari citra 2 dan menyimpannya dalam list dataA
        dataA = [copyCitra2[baris-1, kolom-1], copyCitra2[baris-1, kolom], copyCitra2[baris-1, kolom+1], \
                 copyCitra2[baris, kolom-1], copyCitra2[baris, kolom], copyCitra2[baris, kolom+1], \
                 copyCitra2[baris+1, kolom-1], copyCitra2[baris+1, kolom], copyCitra2[baris+1, kolom+1]]
        
        # Mengurutkan dataA secara ascending
        dataA.sort()
        
        # Mengambil nilai median (elemen ke-4 setelah diurutkan) dan menyimpannya dalam output2
        output2[baris, kolom] = dataA[4]
```

Plot Citra Input dan Output Hasil dari Filter Rerata

```python
# Membuat subplots dengan ukuran 2x2
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()

# Menampilkan citra pertama pada axis pertama
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Input Citra 1")

# Menampilkan citra kedua pada axis kedua
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Input Citra 2")

# Menampilkan output citra pertama pada axis ketiga
ax[2].imshow(output1, cmap='gray')
ax[2].set_title("Output Citra 1")

# Menampilkan output citra kedua pada axis keempat
ax[3].imshow(output2, cmap='gray')
ax[3].set_title("Output Citra 2")

# Mengatur tata letak subplot
fig.tight_layout()
```

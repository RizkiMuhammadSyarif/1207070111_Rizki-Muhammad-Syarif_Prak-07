Konvolusi

Import Library

```python
# Mengimpor library yang diperlukan
# digunakan untuk membuat plot grafik
import matplotlib.pyplot as plt
# digunakan untuk menampilkan plot di dalam notebook
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 

# digunakan untuk operasi array dan manipulasi data numerik
import numpy as np

# digunakan untuk pemrosesan gambar dan komputer visi
import cv2
```

Load Image

```python
# Membaca citra "Shikamaru.jpeg" dan menyimpannya ke dalam variabel citra1
citra1 = imread(fname="Shikamaru.jpeg")

# Menampilkan dimensi citra1 menggunakan atribut shape
print(citra1.shape)

# Menampilkan citra1 dengan colormap 'gray'
plt.imshow(citra1, cmap='gray')
```

Proses Konvolusi

```python
# Definisikan kernel untuk operasi filtering
kernel = np.array([[-1, 0, -1],
                   [0, 4, 0],
                   [-1, 0, -1]])

# Lakukan operasi filtering pada citra1 menggunakan cv2.filter2D
citraOutput = cv2.filter2D(citra1, -1, kernel)

# Menampilkan citra input dan citra output
fig, axes = plt.subplots(1, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Citra Input")

ax[1].imshow(citraOutput, cmap='gray')
ax[1].set_title("Citra Output")
```

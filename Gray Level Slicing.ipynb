Gray Level Slicing

Import Library

```python
# Mengimpor library yang diperlukan
# digunakan untuk pemrosesan gambar dan komputer visi
import cv2
# digunakan untuk operasi array dan manipulasi data numerik
import numpy as np
# digunakan untuk membuat plot grafik
import matplotlib.pyplot as plt
# digunakan untuk menampilkan plot di dalam notebook
%matplotlib inline
```

Gray Level Slice

```python
# Mendefinisikan path direktori gambar
image_path = "Shikamaru.jpeg"

# Membaca gambar menggunakan OpenCV
img = cv2.imread(image_path)

# Mendapatkan dimensi gambar
row, column, _ = img.shape

# Membuat gambar baru dengan ukuran yang sama dengan gambar asli, namun dengan tipe data uint8
img1 = np.zeros((row, column), dtype='uint8')

# Menentukan rentang nilai minimum dan maksimum
min_range = 10
max_range = 60

# Melakukan operasi thresholding pada setiap piksel gambar
for i in range(row):
    for j in range(column):
        if img[i, j, 0] > min_range and img[i, j, 0] < max_range:
            img1[i, j] = 255
        else:
            img1[i, j] = 0
```

Plot Image

```python
# Membuat figure dan axes dengan ukuran 2x2
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

# Menampilkan citra input dan histogramnya
ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title("Citra Input")
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title('Histogram Input')

# Menampilkan citra output setelah thresholding dan histogramnya
ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')

# Menyusun tampilan plot secara rapi
fig.tight_layout()
```

Import Library

```python
# Mengimpor library yang diperlukan
# digunakan untuk membuat plot grafik
import matplotlib.pyplot as plt
# digunakan untuk menampilkan plot di dalam notebook
%matplotlib inline

# digunakan untuk membaca gambar dari file menggunakan library scikit-image (skimage)
from skimage.io import imread
# digunakan untuk mengubah gambar berwarna (RGB) menjadi gambar skala keabuan (grayscale)
from skimage.color import rgb2gray
# modul ini berisi fungsi-fungsi untuk membaca dan menulis gambar dari berbagai format file
from skimage import io, color 
# digunakan untuk operasi array dan manipulasi data numerik
import numpy as np
```

Load & Plot Input Image

```python
# Membaca citra pertama (citra "Shikamaru.jpeg")
citra1 = imread(fname="Shikamaru.jpeg")

# Membaca citra kedua (citra "Random.jpg")
citra2 = imread(fname="Random.jpg")

# Menampilkan shape dari citra pertama dan citra kedua
print('Shape citra 1 : ', citra1.shape)
print('Shape citra 2 : ', citra2.shape)

# Membuat figure dan axes dengan ukuran 1x2
fig, axes = plt.subplots(1, 2, figsize=(10, 10))
ax = axes.ravel()

# Menampilkan citra pertama
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Citra 1")

# Menampilkan citra kedua
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Citra 2")
```

Menyiapkan variable output

```python
# Membuat salinan citra pertama dengan tipe data float
copyCitra1 = citra1.copy().astype(float)

# Membuat salinan citra kedua dengan tipe data float
copyCitra2 = citra2.copy().astype(float)

# Mendapatkan ukuran citra pertama
m1, n1 = copyCitra1.shape[:2]

# Membuat array kosong dengan ukuran citra pertama
output1 = np.empty((m1, n1))

# Mendapatkan ukuran citra kedua
m2, n2 = copyCitra2.shape[:2]

# Membuat array kosong dengan ukuran citra kedua
output2 = np.empty((m2, n2))

# Menampilkan shape dari salinan citra pertama dan output citra pertama
print('Shape copy citra 1 : ', copyCitra1.shape)
print('Shape output citra 1 : ', output1.shape)
print('m1 : ', m1)
print('n1 : ', n1)
print()

# Menampilkan shape dari salinan citra kedua dan output citra kedua
print('Shape copy citra 2 : ', copyCitra2.shape)
print('Shape output citra 2 : ', output2.shape)
print('m2 : ', m2)
print('n2 : ', n2)
print()
```

Proses Filter Rerata Pada Citra Input 1

```python
# Melakukan perulangan untuk setiap baris pada citra pertama
for baris in range(1, m1-1):
    # Melakukan perulangan untuk setiap kolom pada citra pertama
    for kolom in range(1, n1-1):
        # Menghitung jumlah pixel pada citra pertama dengan melakukan penjumlahan pada area 3x3 sekitar pixel saat ini
        jumlah = np.sum(copyCitra1[baris-1:baris+2, kolom-1:kolom+2])
        # Menghitung nilai rata-rata dari jumlah pixel dengan membaginya dengan 9
        output1[baris, kolom] = (1/9) * jumlah
```

Proses Filter Rerata Pada Citra Input 2

```python
# Melakukan perulangan untuk setiap baris pada citra kedua
for baris in range(1, m2-1):
    # Melakukan perulangan untuk setiap kolom pada citra kedua
    for kolom in range(1, n2-1):
        # Menghitung jumlah pixel pada citra kedua dengan melakukan penjumlahan pada area 3x3 sekitar pixel saat ini
        jumlah = np.sum(copyCitra2[baris-1:baris+2, kolom-1:kolom+2])
        # Menghitung nilai rata-rata dari jumlah pixel dengan membaginya dengan 9
        output2[baris, kolom] = (1/9) * jumlah
```

Plot Citra Input dan Output Hasil dari Filter Rerata

```python
# Membuat subplot dengan 2 baris dan 2 kolom
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()

# Menampilkan citra pertama pada subplot pertama
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Input Citra 1")

# Menampilkan citra kedua pada subplot kedua
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Input Citra 2")

# Menampilkan output citra pertama pada subplot ketiga
ax[2].imshow(output1, cmap='gray')
ax[2].set_title("Output Citra 1")

# Menampilkan output citra kedua pada subplot keempat
ax[3].imshow(output2, cmap='gray')
ax[3].set_title("Output Citra 2")
```
